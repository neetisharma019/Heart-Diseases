```python
###Preparing Data Model
```

```python
from sklearn.model_selection import train_test_split  # scaling so that model can understand 120, 67 terms like these
from sklearn.preprocessing import StandardScaler
StandardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol','thalach', 'oldpeak']
df[columns_to_scale] = StandardScaler.fit_transform(df[columns_to_scale])
df.head()
```

```python
x = df.drop(['target'],axis=1) #train all data for x axis except the target data which will be Y axis
y = df['target']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 40)
```

```python
print('x_train:', x_train.size)
print('x_test:', x_test.size)
print('y_train:',  y_train.size)
print('y_test:',y_test.size)
```

```python
#Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model1 = lr.fit(x_train, y_train) #here we're training x and y to predict testing data for x
prediction1 = model1.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, prediction1)    #we used cm to predict our accuracy for tested target data with our tested dataset
cm
```

```python
sns.heatmap(cm, annot = True) #36 TP and 48 TN
```

```python
sns.heatmap(cm, annot = True, cmap = 'BuPu')
```

```python
TP = cm[0][0]
TN = cm[1][1]
FP = cm[0][1]
FN = cm[1][0]
print("testing accuracy :",(TP+TN)/(TP+TN+FP+FN))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction1))
```

```python
#Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
model2 = dtc.fit(x_train, y_train)
prediction2 = model2.predict(x_test)
cm2 = confusion_matrix(y_test, prediction2)    #we used cm to predict our accuracy for tested target data with our tested dataset
cm2
```

```python
sns.heatmap(cm2, annot = True)
```

```python
accuracy_score(y_test, prediction2)
```

```python
print(classification_report(y_test, prediction2))
```
